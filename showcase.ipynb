## Trajectory aware outcomes 

If we lift the assumption that the objective of the Elo system is to determine who is more likely to win we can suggest an information bottleneck aims to predict whose on top of the game. We determine this by taking a machine evaluation on the odds of the game given the setting. $ \mathbb{P}_\theta(\text{win}\ |\ z_t)$. Then we account for the whole game by taking the area under the curve and normalizing against time:

$$0 \le \frac{1}{T}\int_0^{T} \mathbb{P}_\theta(\text{win}\ |\ z_t) dt\le 1$$ Where at time $T$ the game has ended. This will be our final bottleneck such that, unlike in the traditional Elo system the difference of two ratings accounts for the expected outcome of the game, in our system the difference coincides with the expected difference in area under the curve of the machine evaluation. 

$\mathbb{P}_\theta(\ \cdot \ |\ z_t) $ is a parametric memoryless evaluation on the win probability given the state that is provided by a reasonable game policy that only takes into account the current setting as not to bias the stronger rated player.
